```python
import pandas as pd
from bs4 import BeautifulSoup
import re
import requests
```

```python
file_path = '~/Downloads/projects_clean.csv'
df = pd.read_csv(file_path)

df
```

```python
count = 0
def get_image_url(project_url):
    global count
    count += 1
    if count % 20 == 0:
        print(count)
        
    try:
        response = requests.get(project_url)
        soup = BeautifulSoup(response.content, 'html.parser')
        hero = soup.find('div', {'class': 'hero'})
        style = hero['style']
    except:
        return
    match = re.search('url\((".+")\)', style)
    if match:
        image_url = match.group(0)[5:-2]
        return image_url
```

```python
count = 0
df['image_url'] = df.apply(lambda x:get_image_url(x['url']), axis=1)
```

```python
df.to_csv('projects_with_images.csv')
```
